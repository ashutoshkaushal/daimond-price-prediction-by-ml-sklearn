```python
import sklearn 
from sklearn import svm,preprocessing
```

```python
a=sklearn.utils.shuffle(a) #now we will shuffle our data so that analysis could be better
```

```python
a
```

```python
x=a.drop("price",axis=1).values#here we need to know what is x ,x is every feature of daimond accept its price
x=preprocessing.scale(x)
```

```python
y=a["price"].values #and y is only the price of the diamonds
```

```python
test_size=200 
```

```python
x_train=x[:-test_size]
y_train=y[:-test_size]
```

```python
x_test=x[-test_size:]
y_test=y[-test_size:]
```

```python
clf=svm.SVR(kernel="linear")
clf.fit(x_train,y_train) #here data is training 
```

```python
clf.score(x_test,y_test) # here now trained data is now going to be checked how good it is
print(clf.score(x_test,y_test)*100)
```

```python
for x,y in zip(x_test,y_test):
    print(f"Model: {clf.predict([x])[0]},Actual:{y}")
```
